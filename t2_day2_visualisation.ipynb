```
df.groupby(['param_marka-pojazdu'])['price_value'].agg(np.mean).plot(kind='bar', figsize=(15,7))
```

```
(
    df
      .groupby(['param_marka-pojazdu'])['price_value']
      .agg([np.mean, np.median, np.size])
      .sort_values(by='size', ascending=False)
      .head(50)
      
).plot(kind='bar', figsize=(15,7), subplots=True);
```

```
## KRAJ POCHODZENIA
(
    df
      .groupby(['param_kraj-pochodzenia'])['price_value']
      .agg([np.mean, np.median, np.size])
      .sort_values(by='size', ascending=False)
      .head(50)
      
).plot(kind='bar', figsize=(15,7), subplots=True);
```

```
## KOLOR
(
    df
      .groupby(['param_kolor'])['price_value']
      .agg([np.mean, np.median, np.size])
      .sort_values(by='mean', ascending=False)
      .head(50)
      
).plot(kind='bar', figsize=(15,7), subplots=True);
```
